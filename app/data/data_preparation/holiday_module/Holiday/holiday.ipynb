## 2025
- Instruction:
  1. Download Calendar Data:
     - Visit the [New York Calendar](https://fastertechnologies.com/us-government-holidays-ics.html)
     - Download the calendar file and save as `new-york.ics`
  2. Run the following code to convert it to csv file.

```python
from icalendar import Calendar
from datetime import datetime
import pandas as pd

with open('new-york.ics', 'r', encoding='utf-8') as f:
    cal = Calendar.from_ical(f.read())

events = []
for component in cal.walk():
    if component.name == "VEVENT":
        event = {
            "summary": str(component.get('summary')),
            "date": component.get('dtstart').dt.strftime('%Y-%m-%d'),
            "description": str(component.get('description')),
        }
        events.append(event)

df_ics  = pd.DataFrame(events)
df_ics.head()
```

```python

url = "https://www.officeholidays.com/countries/usa/new-york/2024"

# Read holiday table from the webpage
tables = pd.read_html(url)
df_web = tables[0]

# Rename columns for clarity
df_web.columns = ["Day", "Date", "Holiday Name", "Type", "Comments"]

# summary
df_web["summary"] = "New York: " + df_web["Holiday Name"] + df_web["Type"].apply(lambda x: f" ({x})" if pd.notnull(x) else "")

# mapping date "Jan 01" → "2024-01-01"
month_map = {
    'Jan': '01', 'Feb': '02', 'Mar': '03', 'Apr': '04',
    'May': '05', 'Jun': '06', 'Jul': '07', 'Aug': '08',
    'Sep': '09', 'Oct': '10', 'Nov': '11', 'Dec': '12'
}

# mapping date "Jan 01" → "2024-01-01"
df_web["date"] = df_web["Date"].apply(lambda x: f"2024-{month_map[x.split()[0]]}-{x.split()[1].zfill(2)}")

# description
df_web["description"] = df_web["Comments"].fillna('') 


df_web_formatted = df_web[["summary", "date", "description"]]
```

## Combine and save

```python
df_combined = pd.concat([df_ics, df_web_formatted], ignore_index=True)

# Sort by date
df_combined = df_combined.sort_values("date").reset_index(drop=True)

# Save to CSV
df_combined.to_csv("combined_ny_holidays.csv", index=False)
```
