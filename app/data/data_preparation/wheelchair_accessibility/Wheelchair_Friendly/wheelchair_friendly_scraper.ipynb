# Wheelchair Friendly Restaurant
- Instruction:
Run the following code to scrape data from [Wheelchair Friendly Restaurant Finder](https://nyc.maps.arcgis.com/apps/instant/nearby/index.html?appid=064965d88c194077a158e33a7687e296&findSource=-1&find=manhattan&level=12&filter=%7B%22type%22%3A%22select%22%2C%22layerId%22%3A%22192304d2c02-layer-4%22%2C%22expressionId%22%3A%221727388389309%22%2C%22selectedFields%22%3A%5B%22Manhattan%22%5D%7D)


```python
import requests
import pandas as pd

url = "https://services6.arcgis.com/yG5s3afENB5iO9fj/arcgis/rest/services/Wheelchair_Friendly_Restaurant_PROD_view/FeatureServer/0/query"
params = {
    "where": "1=1",
    "outFields": "*",
    "f": "json"
}

response = requests.get(url, params=params)
data = response.json()

# extract features
records = [f['attributes'] for f in data['features']]
wheelchair_df = pd.DataFrame(records)

manhattan_wheelchair_df = wheelchair_df[wheelchair_df['Borough'] == 'Manhattan'].copy()
# save as CSV
manhattan_wheelchair_df.to_csv("manhattan_wheelchair_friendly_restaurants.csv", index=False)

print(len(manhattan_wheelchair_df),"restaurants in Manhattan are wheel chair friendly.")
```
